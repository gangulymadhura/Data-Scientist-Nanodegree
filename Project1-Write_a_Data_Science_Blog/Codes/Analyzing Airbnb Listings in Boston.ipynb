### Missing value imputation
This part handles the missing values.
1. For categorical and string columns missing values are replaced with "missing" text. This done as we don't know the reason why the values are missing. These are later on encoded into numerical values.


2. For numerical features missing values are imputed with a local imputation approach. In local imputation missing values are imputed with local mean or median. The categorical feature that explains maximum variance of a numeric feature is chosen for defining the locality, this feature is selected based on smallest one way ANOVA F-test pvalue. Each missing value in the numeric column is imputed by the mean within the corresponding class of the categorical feature. This is similar to global imputation where missing values are imputed with overall mean or median value, the difference being here the imputation is done within each class of the selected categorical feature.

### Feature Engineering
Before further exploring the data we create some new features that will be useful for answering the question proposed.

#### Calender data
These are the features that are created from the calendar dataset.

- The `date` feature is transformed into various time elements like weekday, month, year for time-based analysis.
- Booking probability is the probability of a listing to be booked and reflects the popularity of a listing as of the time the data was collected. As calendar data has listing bookings for the next 12 months as of 2016 August, we calculate the proportion of days a listings is booked out of total days as booking rate or booking probability.

```python
# create time columns
cal_df['date_year'] = cal_df['date'].dt.year
cal_df['date_month'] = cal_df['date'].dt.month
cal_df['date_qtr'] = cal_df['date'].dt.quarter
cal_df['date_day'] = cal_df['date'].dt.day
cal_df['date_weekday'] = cal_df['date'].dt.day_name()
cal_df['date_week'] = cal_df['date'].dt.week
cal_df['year_month'] = cal_df['date'].dt.strftime('%Y-%m')

# calculate the availability probability for each listing
def count_t(col):
    return sum(col=='t')

# summarize cal_df
cal_summary = cal_df.groupby(['listing_id']).agg({'price':['sum','median'],'available':['count',count_t]}).reset_index()
cal_summary.columns = ['listing_id','total_price','median_price','total_days','days_available']
cal_summary['mean_price'] = cal_summary['total_price'] /cal_summary['days_available']
cal_summary['available_prob'] = cal_summary['days_available'] /cal_summary['total_days']
cal_summary['booking_prob'] = 1 - cal_summary['available_prob']
```

#### Listings data

- The `amenities` column has all amenities together in one string field. This column is split into individial binary columns for each amenity where a value of 1 denotes presence and 0 denotes absence of an amenity at the listing.


- `amenities_count` is the number of amenities available is also created.


- `host_verifications` is a string field which lists all the platforms where a host is verified. This columns too is split into individial binary columns for each verificarion platform.


- The time gap between `host_since` and `last_scraped` is created as `host_tenure_months`, which is the time the host has spent on Airbnb platform.


- The time gap between `first_review` and `last_scraped` is cerated as `time_since_first_review` and time gap between `last_review` and `last_scraped` is cerated as `time_since_last_review`.
 

```python
import itertools
import pandas as pd

listing_df['amenities_count'] = listing_df['amenities'].apply(lambda x : len(x.split(',')))
listing_df['amenities_split'] = listing_df['amenities'].str.strip('[{}]').str.split(",")

# create list of unique amenities
amenities = list(set(list(itertools.chain.from_iterable(listing_df['amenities_split'].tolist()))))
amenities_cleaned = [x.strip('[""]') for x in amenities]

# removing some values
amenities_cleaned.remove('')
amenities_cleaned.remove('translation missing: en.hosting_amenity_49')
amenities_cleaned.remove('translation missing: en.hosting_amenity_50')

# these amenities will be added as indicator columns in the dataset
# for example column amenity_TV =1 if the listing amenities description has a TV

for col in amenities_cleaned:
        listing_df['amenity_'+col] = listing_df['amenities'].str.contains(col).astype(int)

listing_df.columns = listing_df.columns.str.replace(" ","_")
listing_df.columns = listing_df.columns.str.replace("/","")
listing_df.columns = listing_df.columns.str.replace("(","")
listing_df.columns = listing_df.columns.str.replace(")","")
listing_df.columns = listing_df.columns.str.lower()

# drop field amenities
listing_df.drop('amenities',axis=1,inplace=True)
listing_df.drop('amenities_split',axis=1,inplace=True)

# create list of unique host verifications
listing_df['host_verifications_list'] = listing_df['host_verifications'].str.strip('[""[]]').str.split(",")
listing_df['host_verifications_list'][0]

host_ver = list(set(list(itertools.chain.from_iterable(listing_df['host_verifications_list'].tolist()))))
host_ver = [x.replace("'","") for x in host_ver]
host_ver = [x.strip(" ") for x in host_ver]

# removing some values
host_ver.remove('')

for col in host_ver:
        listing_df['host_ver_'+col] = listing_df['host_verifications'].str.contains(col).astype(int)

# drop field host_verifications
listing_df.drop('host_verifications',axis=1,inplace=True)
listing_df.drop('host_verifications_list',axis=1,inplace=True)

# create some more features

# host tenure as of 2016-09-07
last_scraped = listing_df.last_scraped.max()
listing_df['host_tenure_months'] = (last_scraped - listing_df['host_since'])/pd.Timedelta(days=30)
listing_df['time_since_first_review'] = (last_scraped - listing_df['first_review'])/pd.Timedelta(days=30)
listing_df['time_since_last_review'] = (last_scraped - listing_df['last_review'])/pd.Timedelta(days=30)

# merging in cal_summary with listings_df
listing_df = listing_df.merge(cal_summary,left_on='id',right_on='listing_id')
```
